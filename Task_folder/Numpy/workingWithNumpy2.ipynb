```python
# Scalar multiplication 
vector = np.array([2,3,1])
scalar = 3 

scaled_vector =scalar * vector
print(f"Original vector: {vector}")
print(f"Scalar: {scalar}")
print(f"Scaled vector: {scaled_vector} ")

#Geometric interpretation to 2d 
original_2d = np.array([2,1])
scales = [0.5,1,2,-1]

fig=plt.figure(figsize=(10,6))
ax = fig.add_subplot(1, 1,1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
colors = ['red', 'blue', 'green', 'purple']
for i ,scale in enumerate(scales):
    scaled = scale * original_2d
    plt.quiver(0,0, scaled[0],scaled[1],angles='xy',scale_units ='xy', scale =1 ,color=colors[i], width=0.005, label=f'{scale} *v')

plt.xlim(-4,4)
plt.ylim(-4,4)
plt.grid(True)
plt.title("Scalar Multiplication Effects")
plt.legend()
plt.show()
```

```python
# Properties of scalar multiplication
a = np.array([1,2,3])
c1,c2 = 2,3

#associative: c1(c2*a)= (c1*c2)*a

left = c1 * (c2*a)
right = (c1*c2)*a
print(f"c1(c2*a) = {left}")
print(f"(c1*C2)*a = {right}")

#Distributive over vector addition: c(a+b) = ca + cb
left =c1 * (a+b)
right = c1 * a + c1 * b
print(f"c(a+b) = {left}")
print(f"ca + cb = {right}")
print(f"Distributive: {np.array_equal(left,right)}")
```

```python
#physicist's view: vectors have arrows in space 
vector_a = np.array([3,4])
magnitude = np.linalg.norm(vector_a)
print(f"Vector: {vector_a}")
print(f"Magnitude: {magnitude}")

# Direction (unit vector)
unit_vector = vector_a / magnitude 
print(f"Unit vector(direction): {unit_vector}")

#Visualization
plt.figure(figsize=(8,6))
plt.quiver(0,0,vector_a[0],vector_a[1],angles='xy',scale_units='xy', scale=1, color='red', width=0.005)
plt.quiver(0,0,unit_vector[0],unit_vector[1],angles='xy', scale_units='xy', scale=1 , color ='blue', width=0.003)
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.grid(True)
plt.title("Vector: Magnitude and Direction")
plt.legend(['Original Vector', 'Unit Vector'])
plt.show()
```

```python
# A mathematician perspective 
# vectors are any objects that support addition and scalar multiplication
##Example 1: Polynomials as vectors 
# p(x) = 3x^2 + 2x + 1 represented as [1, 2, 3] (coefficients)
poly1 = np.array([1,2,3])  # 1 + 2x + 3x^2
poly2 = np.array([2,-1,1]) #2 -x + x^2

#vector addition for polynomials 
poly_sum = poly1 + poly2
print("polynomials 1:", poly1) 
print("polynomial 2:",poly2)
print("Sum:",poly_sum)# (1+2)+(2-1)x +(3+1)x^2 = 3+x+4x^2

#Scalar multiplication for polynomials
scalar = 2 
poly_scaled = scalar * poly1 
print(f"2 * Polynomial 1:", poly_scaled)

#example 2: functions as vectors (function space)
#this is more abstract, but functions can be added and scaled 
def func1(x):
    return x**2

def func2(x):
    return 2*x

def func_sum(x):
    return func1(x) + func2(x)


x_vals = np.linspace(0,3,100)
plt.figure(figsize=(10,6))
plt.plot(x_vals, func1(x_vals), label='f(x)=x^2')
plt.plot(x_vals, func2(x_vals), label='g(x)=2x')
plt.plot(x_vals,func_sum(x_vals), label='(f+g)(x) = x^2 + 2x')
plt.legend()
plt.title("Function as Vectors")
plt.grid(True)
plt.show()
```

```python
# Audio signals as vectors
sample_rate = 44100
duration = 0.1 # 0.1 seconds
t = np.linspace(0,duration, int(sample_rate * duration))


#Two sine waves (audio signals )
signal1 =0.5*np.sin(2*np.pi * 440 * t ) # a note (440 Hz)
signal2 = 0.3 * np.sin(2* np.pi * 880 * t) # A note one octave higher

#vector addition of audio signals
mixed_signal = signal1 + signal2

#scalar multiplication (volume control)
quiet_signal = 0.5 * mixed_signal

plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.plot(t[:1000],signal1[:1000])
plt.title("signal 1 (440 Hz)")
plt.grid(True)


plt.subplot(2,2,2)
plt.plot(t[:1000],signal2[:1000])
plt.title("signal 2 (880 Hz)")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(t[:1000],mixed_signal[:1000])
plt.title("Mixed signal (Addition)")
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(t[:1000],quiet_signal[:1000],'red')
plt.title(" Quiet signal (Scalar multiplication)")
plt.grid(True)
```

Matrix Operations

    **Matrix Addition**
    
        **just like vector addition, you add each element in the same position.

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[4, 3], [2, 1]])
print("A + B =\n", A + B)
```

```python
# Matrix addition - element-wise operation
A = np.array([[1, 2], 
              [3, 4]])
B = np.array([[5, 6], 
              [7, 8]])

# Addition
C = A + B
print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nA + B:")
print(C)
```

```python
# Properties of matrix addition 
# Commutative: A +B = B+ A
print(f"\nCommutative Property")
print(f"A + B = B + A:{np.array_equal(A+B,B+A)}")
```

```python
# Associative: (A + B) + C = A + (B + C)
C_new = np.array([[1, 1], [1, 1]])
left = (A + B) + C_new
right = A + (B + C_new)
print(f"Associative property:")
print(f"(A + B) + C = A + (B + C): {np.array_equal(left, right)}")
```

```python
# Zero matrix (additive identity)
zeros_matrix = np.zeros_like(A)
print(f"\nAdditive identity:")
print(f"A + 0 = A: {np.array_equal(A + zeros_matrix,A)}")

#Different sized matrces cannot be added 
try:
    D = np.array([[1,2,3]])
    result = A + D # This will fail
except ValueError as e:
    print(f"\nError adding different sized matrices: {e}")
```

**Matrix Multiplication**
- Used to apply transformations, combine features, or calculate neural net outputs.
  - **Rules**
   - A must have same columns as B’s rows.
   - Multiply row-by-column.

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[2, 0],
              [1, 2]])

result = A @ B  # or np.dot(A, B)
print("A x B =\n", result)
```

```python
# Matrix multiplication - not element-wise!
# Rule: (mxk) x(kxn) -> (mxn)

# Compatible matrices
A = np.array([[1, 2, 3],    # 2×3 matrix
              [4, 5, 6]])
B = np.array([[7, 8],       # 3×2 matrix
              [9, 10],
              [11, 12]])
```

```python

# Matrix multiplication using @ operator or np.dot()..
C = A @ B  # or np.dot(A, B)
print("Matrix A (2×3):")
print(A)
print("\nMatrix B (3×2):")
print(B)
print("\nA @ B (2×2):")
print(C)
```

```python
# Manual calculation for understanding
print("\nManual calculation:")
print(f"C[0,0] = {A[0,0]}×{B[0,0]} + {A[0,1]}×{B[1,0]} + {A[0,2]}×{B[2,0]} = {A[0,0]*B[0,0] + A[0,1]*B[1,0] + A[0,2]*B[2,0]}")
print(f"C[0,1] = {A[0,0]}×{B[0,1]} + {A[0,1]}×{B[1,1]} + {A[0,2]}×{B[2,1]} = {A[0,0]*B[0,1] + A[0,1]*B[1,1] + A[0,2]*B[2,1]}")
```

```python
# Matrix multiplication is NOT commutative
# B @ A would be (3×2) @ (2×3) = (3×3)
BA = B @ A
print(f"\nB @ A (3×3):")
print(BA)
print(f"A @ B shape: {C.shape}")
print(f"B @ A shape: {BA.shape}")
print("Matrix multiplication is NOT commutative!")
#nice
```

```python

# Incompatible matrices
try:
    D = np.array([[1, 2]])  # 1×2 matrix
    result = A @ D  # 2×3 @ 1×2 - incompatible !!!!!
except ValueError as e:
    print(f"\nError with incompatible matrices:-->>>{e}")
```
