```python
from textblob import TextBlob

def get_sentiment(sentense):
    polarity = TextBlob(sentense).sentiment.polarity
    if polarity>0:
        return 'positive'
    elif polarity<0:
        return 'negative'
    else:
        return 'neutral'
    

sales['sentiment']=sales['review_content'].apply(get_sentiment)
```
